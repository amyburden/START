```python
import json
import os,glob
from collections import defaultdict
import requests
from io import BytesIO
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import shutil
def load_bad_item_from_box(bb_car):
    bad_item = []
    for item in bb_car:
        if len(bb_car[item]) == 1 and bb_car[item][0] == 'error':
            bad_item.append(item)
    bad_item.sort()
    
    bad_item = [item.split('.')[0] for item in bad_item]
    bad_item = [item.split('_')[0] for item in bad_item]
    bad_item = [item.split('/')[-1] for item in bad_item]
    bad_item = list(set(bad_item))
    return bad_item
```

```python
import base64
import cStringIO
from PIL import Image
import requests
import os, glob,json
from io import open

def get_base64(image):
    buffer = cStringIO.StringIO()
    image.save(buffer, format="JPEG")
    img_str = base64.b64encode(buffer.getvalue())
    return img_str
def pred(path):
    try:
        image = Image.open(path)
        data = get_base64(image)
        data = {'image':data}
        header = {'content-type' : 'application/x-www-form-urlencoded', }
        url = 'https://aip.baidubce.com/rest/2.0/vis-classify/v1/car?access_token=24.4267357fe5871c208eeaef41caac6e91.2592000.1502866757.282335-9891718'
        r = requests.post(url, data=data, headers = header)
        print r.json()['result'][0]
        return r.json()['result'][0]['name']
    except:
        return None

def lookup(d, key, i):
    if key == 'make+model':
        for item in d:
            if item['pp_brand_id']+' '+item['pp_genre_id'] == i:
                return item['chinese']
    else:
        for item in d:
            if item[key] == i:
                return item['chinese']
```

```python
data_dir = '../'
brands = os.listdir(data_dir)
brands.remove('bba')
val_paths = []
val_models = []
val_makes = []
for brand in brands:
    genres = os.listdir(os.path.join(data_dir, brand))
    for genre in genres:
        pics = glob.glob(os.path.join(data_dir, brand, genre)+'/*_b.jpg')
        split = len(pics)*4/5
        #validation data
        for pic in pics[split:]:
            # two pic in a row
            val_paths += [pic[:-6] + '.jpg' + ' ' + pic]
            # label|
            val_models += [brand + ' ' +genre]
            val_makes += [brand]

#indexing building
#l2i_makes = sorted(list(set(train_makes + ['-1'])))
#l2i_models = sorted(list(set(train_models + ['-1'])))
##genere list
genre_list = [json.loads(line) for line in open('./genre_list.json')][0]
bba_info = [json.loads(line) for line in open('./allcar.json',encoding='gbk')]
```

```python
right = 0
wrong_item = []
for i in range(len(val_paths)):
    try:
        make = lookup(bba_info, 'pp_genre_id', val_models[i].split(' ')[1])
        model = genre_list[val_models[i].split(' ')[1]]['name']
        pred1 = pred(val_paths[i].split(' ')[0])
        if (make+model)==pred1:
            right += 1
        else:
            wrong_item.append([make+model, pred1, val_paths[i]])
        if i % 500 == 0:
            print i
    except:
        pass
```
