```python
start_2_yiche = defaultdict(list)
for item in start_car_maping:
    found = 0 
    for k in yiche_car_mapping:
        c = k[2]
        if c == item[3] or c == item[2]+item[3]:
            start_2_yiche[item[1]] = [k[0], k[1], k[2]]
            print item[2],item[3],c
            found = 1
            break
    for k in yiche_car_mapping:
        c = k[2]
        if found == 1:
            break
        if  c in item[2]+item[3] or c in item[2]+' '+item[3] or item[2]+item[3] in c:
            start_2_yiche[item[1]] = [k[0], k[1], k[2]]
            print item[2],item[3], c
            found = 1
            break
    for k in yiche_car_mapping:
        c = k[2]
        if found == 1:
            break
        if found == 0 and len(item[3])>3 and item[3] in c:
            start_2_yiche[item[1]] = [k[0], k[1], k[2]]
            print item[2],item[3], c
            found = 1
            break
    for k in yiche_car_mapping:
        c = k[2]
        if found == 1:
            break
        if found == 0 and item[3] in c:
            start_2_yiche[item[1]] = [k[0], k[1], k[2]]
            print item[2],item[3], c
            found = 1
            break
    if found == 0:
        start_2_yiche[item[1]] = []
        print item[2],item[3],'not found'
```
