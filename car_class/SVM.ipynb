```python
from sklearn import svm
import numpy as np
f = np.load('./yiche_image/map/train.npz')
train_X = f['feat']
train_y = f['label']
assert train_X.shape[0]==train_y.shape[0]
f = np.load('./yiche_image/map/val.npz')
val_X = f['feat']
val_y = f['label']
assert val_X.shape[0]==val_y.shape[0]
```

```python
clf = svm.SVC(max_iter=5000)
```

```python
clf.fit(train_X, train_y)
```

```python
p = clf.predict(val_X)
```

```python
print 1.0*np.sum(p==val_y)/len(val_y)
```
